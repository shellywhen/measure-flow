```python
import pandas as pd
df = pd.read_csv('marieboucher.csv')
table = df.values
```

```python
table[0]
```

```python
nodeCount = 0
node = set([])
nodes = []
for row in table:
    source = row[0]
    target = row[3]
    if source in node:
        idx = nodes.index(source)
        row[0] = idx
    else:
        node.add(source)
        nodes.append(source)
        row[0] = nodeCount
        nodeCount += 1
    if target in node:
        idx = nodes.index(target)
        row[3] = idx
    else:
        node.add(target)
        nodes.append(target)
        row[3] = nodeCount
        nodeCount += 1
            
```

```python
table[0]
```

```python
nodeTable = []
val = 0
for idx, name in enumerate(nodes):
    if(idx > 100):
        val = 1
    nodeTable.append([idx, name, val])
```

```python
df= pd.DataFrame(nodeTable)
```

```python
df.to_csv('Marie Boucher Nodetable.csv', index=False)
```
